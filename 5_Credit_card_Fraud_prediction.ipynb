Import the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#loading the dataset to pandas
credit_card_data=pd.read_csv('/content/creditcard_2023.csv')
```

```python
#printing the first five rows of the dataset
credit_card_data.head()
```

```python
#printing the last five rows of the dataset
credit_card_data.tail()
```

```python
credit_card_data.shape
```

```python
#checking missing values in the dataset
credit_card_data.isnull().sum()
```

```python
#Dropping the missing rows of the dataset
credit_card_data=credit_card_data.dropna()
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

0 represent Legit Transcation

1 represent Fraudulent Transcation

This is highly imbalanced dataset

Separting the legit and fraudulent transcation

```python
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statiscial measure of the data
legit.Amount.describe()
```

```python
#statiscial measure of the data
fraud.Amount.describe()
```

```python
#compare the values of both transcation
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Building the sample dataset containing Similar distribution of legit and fraudulent transcation

No of fradulent transcation--> 212

```python
legit_sample=legit.sample(n=212)
print(legit_sample.shape)
```

Concadenate newly created legit_sample and previously had fraudlent sample

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
#compare the values of both transcation
new_dataset.groupby('Class').mean()
```

```python
#split the data into features and target
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X)
print(Y)
```

Split into training data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=1)
```

Model Training by Logistic Regression Model

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

Model Evalution

```python
#accuracy score on the training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy Score on the training data:',training_data_accuracy)
```

```python
#accuracy score on the test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy Score on the test data:',test_data_accuracy)
```
